```python
import matplotlib.pyplot as plt
def read_log(filename):
    fp=open(filename)
    k=1
    val_result=[]
    test_result=[]
    for line in fp.readlines():
        # train_loss=line[-27:-18]
        if k>543 and k<842 and k%3==1:
         valid=float(line.split("valid:[")[1][0:5])
         test=float(line.split("test:[")[1][0:5])
         val_result.append(valid)
         test_result.append(test)
         k+=1
        # with open('test.txt','a') as fp2:
        #     fp2.write(train_loss+train_acc)
        else:
            k+=1
            continue

    fp.close()
    
    plt.figure(figsize=(8,6))
    x=[i for i in range(1,101)]
    plt.plot(x,val_result,'',label="val")
    plt.plot(x,test_result,'',label="test")
    plt.title('acc')
    plt.legend(loc='upper right')
    plt.xlabel('iter')
    plt.ylabel('')
    plt.grid(x)
    plt.show()

def main():
    filename='1_hiv_scaffold.log'
    read_log(filename)

if __name__=='__main__':
    main()
```
